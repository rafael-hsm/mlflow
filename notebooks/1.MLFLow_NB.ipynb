```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import *
import matplotlib.pyplot as plt

import mlflow
import mlflow.sklearn
```

```python
url = "https://raw.githubusercontent.com/rafael-hsm/mlflow/main/data/Credit.csv"
credito = pd.read_csv(url)
credito.shape
```

```python
credito.head()
```

```python
for col in credito.columns:
    if credito[col].dtype == 'object':
        credito[col] = credito[col].astype('category').cat.codes
```

```python
credito.head()
```

```python
previsores = credito.iloc[:,0:20]
classe = credito.iloc[:,20]
```

```python
previsores
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(
    previsores,
    classe,
    test_size=0.3,
    random_state=123
    )
```

```python
naive_bayes = GaussianNB()
naive_bayes.fit(X_treinamento, y_treinamento)
previsoes = naive_bayes.predict(X_teste)

acuracia = accuracy_score(y_teste, previsoes)
print(acuracia)
```

```python
recall = recall_score(y_teste, previsoes)
precision = precision_score(y_teste, previsoes)
```

```python
import mlflow
import mlflow.sklearn
from mlflow.models import infer_signature
from sklearn.metrics import (
    accuracy_score, recall_score, precision_score, f1_score,
    roc_auc_score, log_loss, ConfusionMatrixDisplay, RocCurveDisplay
)
import matplotlib.pyplot as plt

mlflow.set_experiment("nbexperimento")

with mlflow.start_run():
    naive_bayes = GaussianNB()
    naive_bayes.fit(X_treinamento, y_treinamento)

    # preds e probas
    previsoes = naive_bayes.predict(X_teste)
    probas = naive_bayes.predict_proba(X_teste)[:, 1]  # classe positiva

    # métricas
    acuracia  = accuracy_score(y_teste, previsoes)
    recall    = recall_score(y_teste, previsoes)
    precision = precision_score(y_teste, previsoes)
    f1        = f1_score(y_teste, previsoes)
    auc       = roc_auc_score(y_teste, probas)   # usa probas!
    log       = log_loss(y_teste, probas)        # idem

    # registrar métricas
    mlflow.log_metric("acuracia", acuracia)
    mlflow.log_metric("recall", recall)
    mlflow.log_metric("precision", precision)
    mlflow.log_metric("f1", f1)
    mlflow.log_metric("auc", auc)
    mlflow.log_metric("log", log)

    # gráficos
    ConfusionMatrixDisplay.from_estimator(naive_bayes, X_teste, y_teste)
    plt.title("Matriz de Confusão")
    plt.savefig("confusion.png"); plt.close()

    RocCurveDisplay.from_estimator(naive_bayes, X_teste, y_teste)
    plt.title("Curva ROC")
    plt.savefig("roc.png"); plt.close()

    # logar gráficos
    mlflow.log_artifact("confusion.png")
    mlflow.log_artifact("roc.png")

    # log do modelo (sem warning) + assinatura
    signature = infer_signature(X_teste, previsoes)
    mlflow.sklearn.log_model(
        sk_model=naive_bayes,
        name="ModeloNB",                 # use name=
        input_example=X_teste.iloc[:5],  # ajuda a inferir schema
        signature=signature
    )

    # informações da execução
    print("Run ID:", mlflow.active_run().info.run_id)

mlflow.end_run()
```
